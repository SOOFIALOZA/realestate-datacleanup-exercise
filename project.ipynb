#### Exercise 10. Is the average of "Valdemorillo" and "Galapagar" price per square meter (price/m2) the same? (★★☆)

Print the both average prices and then write a conclusion about

Hint: Create a new column called `pps` (price per square) and then analyse the values

```python
# TODO
import pandas as pd
import numpy as np
ds = pd.read_csv('assets/real_estate.csv', sep=';')
# Filter the data for "Valdemorillo" and "Galapagar"
valdemorillo = ds[ds["level5"] == "Valdemorillo"]
galapagar = ds[ds["level5"] == "Galapagar"]

# Create a new column "pps" (price per square) by dividing "price" by "surface"
valdemorillo["pps"] = valdemorillo["price"] / valdemorillo["surface"]
galapagar["pps"] = galapagar["price"] / galapagar["surface"]

# Calculate the average "pps" for "Valdemorillo" and "Galapagar"
avg_pps_valdemorillo = valdemorillo["pps"].mean()
avg_pps_galapagar = galapagar["pps"].mean()

# vemos los valores del pps
print("Average Price per Square Meter (Valdemorillo):", avg_pps_valdemorillo)
print("Average Price per Square Meter (Galapagar):", avg_pps_galapagar)

#comprobbar que dan el mismo valoe 
if avg_pps_valdemorillo == avg_pps_galapagar:
    print("The average price per square meter is the same for Valdemorillo and Galapagar.")
else:
    print("The average price per square meter is different for Valdemorillo and Galapagar.")

```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in the question 14) (★★★)

For the normalization method you can use the one you consider to, there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot

Hint: You can help yourself reviewing the multihist demo of Matplotlib

```python
# TODO

import pandas as pd
import matplotlib.pyplot as plt
ds = pd.read_csv('assets/real_estate.csv', sep=';')
filtercitys = ds["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])
dsfiltrado = ds[filtercitys]

fig, ax = plt.subplots()

city_data = []

for city in ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]:
    city_prices = dsfiltrado[dsfiltrado["level5"] == city]["price"]

    z_scores = (city_prices - city_prices.mean()) / city_prices.std()

    city_data.append(z_scores)

plt.hist(city_data, bins=20, label=["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])
plt.legend()
plt.xlabel("Normalized Prices")
plt.ylabel("Frequency")
plt.title("Normalized Prices Distribution for Selected Cities")
plt.show()
```

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
# TODO

import pandas as pd
import matplotlib.pyplot as plt

# Cargar el conjunto de datos
ds = pd.read_csv('assets/real_estate.csv', sep=';')

# Filtrar las ciudades de interés
filter_cities = ds["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])
dsfiltrado = ds[filter_cities]

# Crear una figura con subtramas para organizar los gráficos
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Crear un gráfico de dispersión para cada población en su subtrama correspondiente
for i, city in enumerate(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]):
    row, col = i // 2, i % 2
    city_data = dsfiltrado[dsfiltrado["level5"] == city]
    
    # Scatter plot de price vs. surface
    axes[row, col].scatter(city_data["surface"], city_data["price"], label=city, alpha=0.5)
    
    axes[row, col].set_title(f"{city} Population")
    axes[row, col].set_xlabel("Surface")
    axes
```
